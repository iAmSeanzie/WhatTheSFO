## Average Prices by Neighborhood

In this section, you will use hvplot to create two interactive visulizations of average prices with a dropdown selector for the neighborhood. The first visualization will be a line plot showing the trend of average price per square foot over time for each neighborhood.  The second will be a line plot showing the trend of average montly rent over time for each neighborhood.

**Hint:** It will be easier to create a new DataFrame from grouping the data and calculating the mean prices for each year and neighborhood

```python
# Group by year and neighborhood and then create a new dataframe of the mean values
year_hood_grp = (
#    sfo_data[[sfo_data.index, "year", "neighborhood"]],
    sfo_data.groupby([sfo_data.index, "neighborhood"])
    .mean()
    .reset_index()
)
year_hood_grp.head(10)
```

```python
# Use hvplot to create an interactive line chart of the average price per sq ft.
# The plot should have a dropdown selector for the neighborhood
year_hood_grp.hvplot.line(
    x = "year",
    y = "sale_price_sqr_foot",
    xlabel = "Year",
    ylabel = "Avg. Sale Price per Square Foot",
    groupby = "neighborhood")
```

```python
# Use hvplot to create an interactive line chart of the average monthly rent.
# The plot should have a dropdown selector for the neighborhood
year_hood_grp.hvplot.line(
    x = "year",
    y = "gross_rent",
    xlabel = "Year",
    ylabel = "Avg. Sale Price per Square Foot",
    groupby = "neighborhood")
```

## Neighborhood Map

In this section, you will read in neighborhoods location data and build an interactive map with the average house value per neighborhood. Use a `scatter_mapbox` from Plotly express to create the visualization. Remember, you will need your Mapbox API key for this.

### Load Location Data

```python
# Load neighborhoods coordinates data
ld = pd.read_csv(Path("Data/neighborhoods_coordinates.csv")).drop_duplicates()
data_to_plot = ld[["Neighborhood", "Lat", "Lon"]]
#filtered_data = ld[ld["StateDesc"] == "California"]
```

### Data Preparation

You will need to join the location data with the mean values per neighborhood.

1. Calculate the mean values for each neighborhood.

2. Join the average values with the neighborhood locations.

```python
# Calculate the mean values for each neighborhood

values = sfo_data.groupby("neighborhood").mean().rename(index={"neighborhood": "Neighborhood"})
values.head()
```

```python
# These are the correct answers
#	Neighborhood	sale_price_sqr_foot	housing_units	gross_rent
#0	Alamo Square	366.020712	378401.0	2817.285714
#1	Anza Vista 	373.382198	379050.0	3031.833333
#2	Bayview	204.588623	376454.0	2318.400000
#3	Bayview Heights	590.792839	382295.0	3739.000000
#4	Bernal Heights	576.746488	379374.5	3080.333333
```

```python
# Join the average values with the neighborhood locations
joined = values.merge([ld["Neighborhood"]='neighborhood'], join='inner')
joined
```

### Mapbox Visualization

Plot the average values per neighborhood using a Plotly express `scatter_mapbox` visualization.

```python
# Set the mapbox access token
load_dotenv()
map_box_api = os.getenv("mapbox")
px.set_mapbox_access_token(map_box_api)
# Create a scatter mapbox to analyze neighborhood info
map_plot = px.scatter_mapbox(
    joined,
    lat="Lat",
    lon="Lon",
    size="sale_price_sqr_foot",
    color="neighborhood",
    zoom=4
)
```
